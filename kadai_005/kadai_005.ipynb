```python
import matplotlib.pyplot as plt
import matplotlib.style
matplotlib.style.use("default")
import numpy as np

fig, axes = plt.subplots(2,3,tight_layout=True)

# 折れ線グラフ
y0 = [1,2,-5,2]
axes[0,0].plot(y0)

# sin関数の曲線
x0 = np.linspace(0,10,100)
y1 = 2 + 2 * np.sin(2 * x0)
axes[0,1].plot(x0,y1, linewidth=2.0)

axes[0,1].set(xlim=(0,10), xticks=np.arange(0,10),
       ylim=(-1,5), yticks=np.arange(-1,6))

# ヒストグラム
np.random.seed(30)
x1 = np.random.normal(15,10,2000)
plt.xlim(-10,40)
axes[0,2].hist(x1, bins=9)


# 散布図
np.random.seed(10)
x2 = 4 + np.random.normal(0,2,60)
y2 = 4 + np.random.normal(0,2,len(x2))
axes[1,0].scatter(x2,y2)

plt.show()
```
